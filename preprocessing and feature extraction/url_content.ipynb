```python
import pandas as pd
import re
from bs4 import BeautifulSoup  # To remove HTML tags
from textblob import TextBlob
import nltk

# Load the dataset
http_df = pd.read_csv("./downsampled_processed_http.csv", low_memory=False)

# Ensure the content column exists
if "content" in http_df.columns:
    
    def preprocess_text(text):
        """Preprocess webpage content before sentiment analysis."""
        if pd.isna(text):  # Handle NaN values
            return ""
        
        text = BeautifulSoup(text, "html.parser").get_text()  # Remove HTML tags
        text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # Remove special characters
        text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
        return text.lower()  # Convert to lowercase

    # Apply preprocessing
    http_df["cleaned_content"] = http_df["content"].astype(str).apply(preprocess_text)

    # Apply sentiment analysis
    http_df["sentiment"] = http_df["cleaned_content"].apply(lambda x: TextBlob(x).sentiment.polarity)

    # Drop the cleaned content column if not needed
    http_df.drop(columns=["cleaned_content"], inplace=True)

    # Save the modified dataset
    http_df.to_csv("http_with_sentiment.csv", index=False)

    print("Sentiment analysis completed. Output saved as http_with_sentiment.csv")

else:
    print("Error: 'content' column not found in http.csv")
```
